***A great way to visualize these variables is by using scatterplots with fitted lines.
In order to start understanding the (linear) relationship between an individual variable and the price. We can do this by using "regplot", which plots the scatterplot plus the fitted regression line for the data.

```python
sns.regplot(x="GDP per capita", y="Overall rank", data=df)  # use seaborn.regplot (sns) to plot a regression line
plt.ylim(0,)                 # the y limit is ylim(0,) 0 to infinity 
```

Observation: As GDP per capita goes up, the Overall rank goes down: this indicates a negative direct correlation between these two variables.
GDP per capita seems like a pretty good predictor of Overall rank since the regression line is almost a perfect diagonal line.

We can examine the correlation between 'GDP per capita' and 'Overall rank' and see it's approximately -0.81 so they are strong negative correlation. 

```python
df[["GDP per capita", "Overall rank"]].corr()  # double brackets for DataFrame
```

####  Correlation: Pearson Correlation
* #### The Pearson Correlation measures the linear dependence between two variables X and Y.
* #### The resulting coefficient is a value between -1 and 1 inclusive, where:
 * #### 1: Total positive linear correlation,   0.6< p < 1 is considered as relatively positive strong correlation
 * #### 0: No linear correlation, the two variables most likely do not affect each other.
 * #### -1: Total negative linear correlation.  -1< p<-0.6 is rlatively strong negative correlation 

```python
def linear_correlation(X, Y):   # define a linear correlation function with two parameters( column names)
    
    print(df[[X, Y]].corr())     # find the correlation for any two columns 
    sns.regplot(x=X, y=Y, data=df)  # use seaborn.regplot (sns) to plot a regression line
    plt.ylim(0,)                        
    
linear_correlation("GDP per capita", "Healthy life expectancy" )
    
```

From the correlation we can see that the two attributes "GDP per capita"and  "Healthy life expectancy" has p=0.83 which indicates that the correlation between them is a strong positive correlation; 

From the regression line we can see thatHealthy life expectancy goes up  as "GDP per capita" grows bigger.
GDP per capita seems like a pretty good predictor of Healthy life expectancy since the regression line is almost a perfect diagonal line.

```python
linear_correlation("Generosity", "Score")
```

Conclusion: Generosity and Score has very weak correlation. People who are generous might be a little more happier but not much!

```python
linear_correlation("Perceptions of corruption", "Score")
```

Conclusion: Perceptions of corruption and Score has positive correlation. People who live in a country with less corruption tend to be a little more happier. 

```python
linear_correlation("Freedom to make life choices","Perceptions of corruption")
```

Conclusion: These two have positive correlation which indicates the country where people have more freedom tend to be less corrupted.
